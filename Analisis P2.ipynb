# ¿Qué horas del día son más rentable para cada una de las tiendas? Razone su respuesta.

```python
query = """ 
with hours as (
	select fecha::time, idsucursal,
	monto_total,
	iva
	from factura
)
select date_trunc('hour', fecha) as hora,
idsucursal,
cast(avg(monto_total) as int) as total,
cast(avg(iva) as int) as total_iva
from hours
group by hora, idsucursal
order by total desc
"""
select(query)
```
